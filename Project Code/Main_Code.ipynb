```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go # or plotly.express as px 
from dash.dependencies import Input, Output
import numpy as np
```

```python
trump_tweet = pd.read_csv("trump_insult_tweets_2014_to_2021.csv")
covid_cases_US = pd.read_csv("all-states-history.csv")
covid_cases_US.isnull().sum()
covid_cases_US.Death = covid_cases_US.Death.fillna(0)
covid_cases_US
```

```python
#external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__)

colors = {
    'background': '#FFFFFF',
    'text': '#000000'}

#data section

#df = df.groupby(['state', 'incident_type', 'fy_declared'])
#df.set_index()
# print(df)

#app layout section
app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    
    html.H1("US Covid Case Surge From Start Until Now", 
            style={'textAlign': 'center',
                   'color': colors['text']}),
        dcc.Dropdown(id = "selected_state",
                options=[
                    {"label": x, "value": x} for x in sorted(covid_cases_US.state.unique())],
                 multi=False,
                 value="AK"
                 ), 
    html.Div(id ='output_container'),
    html.Br(),    
    dcc.Graph(id = 'my_US_covid_graph', figure={})
])

@app.callback(
     Output(component_id='my_US_covid_graph', component_property='figure'),
    [Input(component_id='selected_state', component_property='value')])
def update_graph(option_selected): 
#     print(option_selected)
#     print(type(option_selected))

    df2 = covid_cases_US.copy() #Must use a copy of dataframe
    df2 = df2[df2['state']== option_selected]
    #dff = dff[dff['incident_type']== "Fire"]
#plotly express
    fig = px.line(df2, x="Date", y="Death",color="state", title = "Covid Cases Increases by Date in Different States")

    
#     fig.update_layout(                             #to change graph color, font, size, etc                 
#     plot_bgcolor=colors['background'],
#     paper_bgcolor=colors['background'],
#     font_color=colors['text'],
#     title="US Disaster by State and Year",
#     xaxis_title="Type of Disaster",
#     yaxis_title="Number of Reported Disaster",
#     )
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black')
    fig.update_yaxes(showline=True, linewidth=2, linecolor='black')

    return fig #the return obj will be the output and if there are many output, it will go in order ( 1 obj => 1st output)


if __name__ == '__main__':
    app.run_server(debug=True,use_reloader=False, port=9000)
```

```python
#hate tweet graph

import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go # or plotly.express as px 
from dash.dependencies import Input, Output
import numpy as np

#external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__)

colors = {
    'background': '#FFFFFF',
    'text': '#000000'}

#data section
df = pd.read_csv("ALL_TWEET_SENTIMENT.csv") #no null or duplicate dataset, maybe bc this dataset isn't in the same file as the python file
#df = df.groupby(['state', 'incident_type', 'fy_declared'])
#df.set_index()
# print(df)
df['Count of China Virus'] = df['Text'].str.count('China Virus')
df['Count of fuckchina'] = df['Text'].str.count('fuckchina')
df['Count of Chinacoronavirus'] = df['Text'].str.count('Chinacoronavirus')
df['Count of China Corona Virus'] = df['Text'].str.count('China Corona Virus')
df['Count of Go Back to China'] = df['Text'].str.count('Go Back to China')
df['Count of chinaliedpeopledied'] = df['Text'].str.count('chinaliedpeopledied')
df['Count of Asian Virus'] = df['Text'].str.count('Asian Virus')
df['Count of Wuhan Virus'] = df['Text'].str.count('Wuhan Virus')
df['Count of Chinese Virus'] = df['Text'].str.count('Chinese Virus')
df['Count of Bat Eater'] = df['Text'].str.count('Bat Eater')
df['Count of nukechina'] = df['Text'].str.count('nukechina')
#app layout section
app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    
    html.H1("US Disaster Declarations", 
            style={'textAlign': 'center',
                   'color': colors['text']}),
        dcc.Dropdown(id = "selected_year",
                options=[
                    {"label": x, "value": x} for x in sorted(df['key word'].unique())],
                 multi=False,
                 ), 
    html.Div(id ='output_container'),
    html.Br(),    
    dcc.Graph(id = 'my_disaster_map', figure={})
])

@app.callback(
     Output(component_id='my_disaster_map', component_property='figure'),
    [Input(component_id='selected_year', component_property='value')])
def update_graph(option_selected): 
#     print(option_selected)
#     print(type(option_selected))
    y = str(option_selected)
    dff = df.copy() #Must use a copy of dataframe
    dff = dff[dff['key word']== option_selected]
    #dff = dff[dff['incident_type']== "Fire"]
#plotly express
    fig = px.line(dff, x="Datetime", y=dff['Count of'+' {}'.format(y)], title = "Covid Cases Increases by Date in Different States")
    
#     fig.update_layout(                             #to change graph color, font, size, etc                 
#     plot_bgcolor=colors['background'],
#     paper_bgcolor=colors['background'],
#     font_color=colors['text'],
#     title="US Disaster by State and Year",
#     xaxis_title="Type of Disaster",
#     yaxis_title="Number of Reported Disaster",
#     )
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black')
    fig.update_yaxes(showline=True, linewidth=2, linecolor='black')

    return fig #the return obj will be the output and if there are many output, it will go in order ( 1 obj => 1st output)


if __name__ == '__main__':
    app.run_server(debug=True,use_reloader=False)
```
